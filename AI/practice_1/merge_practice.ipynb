```python
#테이블 병합(merge)하는 작업: 분석하기 위해 만들음.

import pandas as pd
from debugpy.launcher import channel
```

```python
sales_df = pd.read_excel("data/Sales.xlsx", sheet_name="Sheet1")#대소문자 틀리면 오류뜸.
details_df = pd.read_excel("data/Details.xlsx", sheet_name=None)
```

```python
print(details_df.keys())
```

```python
regions_df = details_df["지역"]
promotion_df = details_df["프로모션"]
channel_df = details_df["채널"]
date_df = details_df["날짜"]
customers_df = details_df["2018년도~2022년도 주문고객"]
products_df = details_df["제품"]
categories_df = details_df["분류"]
product_categories_df = details_df["제품분류"]
print(sales_df.keys())
```

```python
merged_df=pd.merge(sales_df, products_df,on="제품코드", how="left")
merged_df

```

```python
merged_df=pd.merge(merged_df, customers_df, on="고객코드", how="left")
merged_df
```

```python
merged_df=pd.merge(merged_df, promotion_df,on="프로모션코드", how="left")
merged_df
```

```python
merged_df=pd.merge(merged_df,channel_df,on="채널코드", how="left")
merged_df
```

```python
merged_df=pd.merge(merged_df, product_categories_df,on="제품분류코드", how="left")
merged_df
```

```python
merged_df=pd.merge(merged_df, regions_df,on="지역코드", how="left")
merged_df.keys()
```

```python
merged_df=pd.merge(merged_df, date_df,on="날짜", how="left")
merged_df.keys()
```
